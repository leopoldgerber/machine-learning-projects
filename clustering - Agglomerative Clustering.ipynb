To create clusters, we need to measure the distance between the data points. The algorithm uses Euclidean distance to do this.

Using single linkage, we look for the pair-wise data points that are closest to each other. We then join them together. This can also be used to merge clusters or to merge data points based on the measured distance between them.

<left>
<div style="color:white;
           display:fill;
           border: 0px;
           border-bottom: 2px solid #AAA;
           font-size:80%;
           letter-spacing:0.5px">
<h2 style="padding: 10px;
           color:#212121;">Cluster distance measures 
</h2>
</div>    
</left>

**Single link:**
$D(c_{1}, c_{2}) = \min_{x_{1}\in c_{1}, x_{2}\in c_{2}} D(x_{1}, x_{2})$
- distance between closest elements in clusters 
- produces long chains 

**Complete link:**
$D(c_{1}, c_{2}) = \max_{x_{1}\in c_{1}, x_{2}\in c_{2}} D(x_{1}, x_{2})$
- distance between farthest elements in clusters 
- forces "spherical" dusters with consistent "diameter" 

**Average link:**
$D(c_{1}, c_{2}) = \frac{1}{|c_{1}|}\frac{1}{|c_{2}|}\sum_{x_{1}\in c_{1}}\sum_{x_{2}\in c_{2}} D(x_{1}, x_{2})$
- average of all pairwise distances 
- less affected by outliers

**Centroids:**
$D(c_{1}, c_{2}) = D \Bigg(\Bigg(\frac{1}{|c_{1}|}\sum_{x\in c_{1}}\overrightarrow{x} \Bigg), \Bigg(\frac{1}{|c_{2}|}\sum_{x\in c_{2}}\overrightarrow{x} \Bigg) \Bigg)$
- distance between centroids (means) of two clusters 

**Ward's method:**
$TD_{c_{1}\cup c_{2}} = \sum_{x\in c_{1}\cup c_{2}}D(x, \mu_{c_{1}\cup c_{2}})^{2}$
- consider joining tiktei rflow does it change the total distance (TD) from centroids?

## Dataset

This data set contains information on user preference data from 73,516 users on 12,294 anime. Each user is able to add anime to their completed list and give it a rating and this data set is a compilation of those ratings.

## Features

-- *Anime.csv* --
- **anime_id** - myanimelist.net's unique id identifying an anime
- **name** - full name of anime
- **genre** - comma separated list of genres for this anime
- **type** - movie, TV, OVA, etc.
- **episodes** - how many episodes in this show. (1 if movie)
- **rating** - average rating out of 10 for this anime
- **members** - number of community members that are in this anime's

-- *Rating.csv* --
- **user_id** - non identifiable randomly generated user id
- **anime_id** - the anime that this user has rated
- **rating** - rating out of 10 this user has assigned (-1 if the user watched it but didn't assign a rating)

```python
anime[['genre', 'type']] = anime[['genre', 'type']].fillna('Unknown')

def split_genre(line):
    splt = [i.replace(',', '').lower() for i in line.split()]
    return splt

anime['genre_dict'] = anime['genre'].apply(split_genre)
#anime['genre_dict'].astype(str).str.replace("'", "").str.strip('[]').str.get_dummies(sep = ', ')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

MLB = MultiLabelBinarizer()

genre_data = MLB.fit_transform(anime['genre_dict'])

genre_labels = MLB.classes_
genre_df = pd.DataFrame(genre_data, columns = genre_labels, index = anime.index)

anime = pd.concat([anime, genre_df], axis = 1)

anime.drop(columns = 'of', inplace = True)
remove_idx = np.argwhere(genre_labels == 'of')
genre_labels = np.delete(genre_labels, remove_idx)
```

```python
top_5_genre = pd.DataFrame(anime[genre_labels].sum().sort_values(ascending = False).head(5)
                          ).reset_index().rename(columns = {'index':'genre', 0:'sum'})

plt.figure(figsize = (14, 5))
sns.catplot(data = top_5_genre, x = 'sum', y = 'genre', kind = 'bar', aspect = 3, color = 'green', alpha = .5)
plt.title('Top-5 wise assigned Genres', fontsize = 16)
plt.subplots_adjust(hspace = .5, wspace = .5)
plt.xlabel('Sum of mentioned genres', fontsize = 11)
plt.ylabel('Genres', fontsize = 11)
plt.show()
```

<div class = 'alert alert-info alert-block'>
    <b>Interpretation:</b>  The most popular genre assigned to anime is comedy.
</div>

```python
top_type = pd.DataFrame(anime.groupby(['type'])['anime_id'].count().sort_values(ascending = False)).reset_index()

plt.figure(figsize = (14, 5))
sns.catplot(data = top_type, x = 'anime_id', y = 'type', kind = 'bar', aspect = 3, color = 'green', alpha = .5)
plt.title('Number of assigned Types', fontsize = 16)
plt.subplots_adjust(hspace = .5, wspace = .5)
plt.xlabel('Sum of Types', fontsize = 11)
plt.ylabel('Types', fontsize = 11)
plt.show()
```

<div class = 'alert alert-info alert-block'>
    <b>Interpretation:</b>  Obvious numbers, since an anime usually contains a large number of episods.
</div>

```python
# Mean rating by user_id
mean_rating = rating.groupby(['user_id'])['rating'].mean().reset_index().rename(columns = {'rating':'mean_rating'})

# Merge together rating and mean rating
users = pd.merge(rating, mean_rating, on = ['user_id','user_id'])

# Drop rows with users rating less then the mean rating of a user
users = users.drop(users[users.rating < users.mean_rating].index)

# Merge anime df and users df
df_merge = pd.merge(anime,users,on=['anime_id','anime_id'])
# Drop rows for a faster performance
df_merge = df_merge[df_merge.user_id <= 1000]

# Create mxn matrix, where m - user_id, n - anime
user_anime = pd.crosstab(df_merge['user_id'], df_merge['name'])
user_anime.head(10)
```

### Agglomerative Clustering

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(user_anime, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

agg_cl = AgglomerativeClustering(n_clusters = 10, affinity = 'euclidean', linkage = 'ward')
predicted = agg_cl.fit_predict(user_anime)
```
